# Widget Debug Test
Run the cell below. If you see a slider and a button, your environment is working.

```python
import ipywidgets as widgets
from IPython.display import display

slider = widgets.IntSlider(description="Test Slider")
button = widgets.Button(description="Click Me")

output = widgets.Output()

def on_click(b):
    with output:
        print(f"Slider Value: {slider.value}")

button.on_click(on_click)

print("Displaying widgets...")
display(slider, button, output)
```
